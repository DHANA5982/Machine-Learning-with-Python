# OOPS method

```python
a = np.linspace(0, 10, 11)
a
```

```python
b = a**3
b
```

```python
x = np.arange(10)
x
```

```python
y = x*2
y
```

```python
fig = plt.figure(figsize=(10, 10))
axes1 = fig.add_axes([0,0,.5,1])
axes1.plot(a, b)
axes2 = fig.add_axes([.5, 0, .5, 1])
axes2.plot(x,y)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
# large pic
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a, b)
axes1.set_xlim(0, 10)
axes1.set_ylim(1, 1000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Exponential')

# small pic
axes2 = fig.add_axes([0.1,.5, .5, .4])
axes2.plot(x,y)
axes2.set_xlim(0, 9)
axes2.set_ylim(0, 18)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Linear')

plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
# large pic
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a, b)
axes1.set_xlim(0, 10)
axes1.set_ylim(1, 1000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Exponential')

# small pic
axes2 = fig.add_axes([0.1,.5, .5, .4])
axes2.plot(a,b)
axes2.set_xlim(2, 8)
axes2.set_ylim(0, 600)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoom in')

plt.show()
```

```python
fig = plt.figure(dpi = 100, figsize = (4, 4))
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a, b)
fig.savefig('new_figure.png', bbox_inches='tight')
```

# Subplots

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(10)
y = x*2
```

```python
fig, ax = plt.subplots(2,2)
ax[0, 0].plot(x, y)
ax[1, 1].plot(a,b)
plt.tight_layout()
plt.show()
```

# Styling

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, label='x vs x')
ax.plot(x, x**2, label='x vs exponential x')
# ax.legend(loc='upper left')
# ax.legend(loc=2)
ax.legend(loc=(1.1, .5))
```

# visual styling

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,x, color='darkblue')
ax.plot(x,x, color='#e6c017', label='x')
ax.plot(x, x+1, color='darkgreen', label='x+1')
ax.legend()
plt.show()
```

# line and markers

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,x, color='darkblue')
ax.plot(x,x, color='#e6c017', label='x', linewidth=0.7, linestyle='--')
ax.plot(x, x+1, color='darkgreen', label='x+1', lw=5, ls='-.')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
line=ax.plot(x,x, color='red', lw=5)
line[0].set_dashes([1,2,1,2,10,2])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x+1, color='black', lw=1, marker='+', ls='--')
ax.plot(x, x, 'rs--', lw=1, ms=10, mfc='k', mew=3, mec='b')
ax.plot(x,x-1, 'bp', ls='--', markersize=10)
plt.show()
```

# Twin Axis

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
# for label in ax1.get_yticklabels():
    # label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```
